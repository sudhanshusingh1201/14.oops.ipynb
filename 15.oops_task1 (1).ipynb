```python
class BankAccount:
    def __init__(self, accountNumber, name, balance):
        self.accountNumber = accountNumber
        self.name = name
        self.balance = balance
        
    def Deposit(self, amount):
        self.balance += amount
        print(f"Deposited: {amount}. New balance: {self.balance}")

    def Withdrawal(self, amount):
        if amount > self.balance:
            print("Insufficient balance!")
        else:
            self.balance -= amount
            print(f"Withdrawn: {amount}. New balance: {self.balance}")

    def bankFees(self):
        fee = self.balance * 0.05  # 5% fee
        self.balance -= fee
        print(f"Bank fees of 5% applied: {fee}. New balance: {self.balance}")

    def display(self):
        print(f"Account Number: {self.accountNumber}")
        print(f"Account Holder: {self.name}")
        print(f"Account Balance: {self.balance}")


# ✅ User input le lo
accNo = int(input("Enter account number: "))
name = input("Enter account holder name: ")
balance = float(input("Enter initial balance: "))

# ✅ BankAccount object banao (variable sahi karo)
acc = BankAccount(accNo, name, balance)

# ✅ Show account details
acc.display()

# ✅ Deposit karo
acc.Deposit(5000)

# ✅ Withdrawal karo
acc.Withdrawal(2000)

# ✅ Bank fees lagao
acc.bankFees()

# ✅ Final details dikhao
acc.display()
```

```python
class Computation:
    def __init__(self):
        # Default constructor — no parameters
        pass

    def Factorial(self, n):
        fact = 1
        for i in range(1, n + 1):
            fact *= i
        print(f"Factorial of {n} is: {fact}")
        return fact

    def naturalSum(self, n):
        total = n * (n + 1) // 2  # Using formula
        print(f"Sum of first {n} natural numbers is: {total}")
        return total

    def testPrime(self, n):
        if n <= 1:
            print(f"{n} is NOT prime.")
            return False
        for i in range(2, int(n**0.5) + 1):
            if n % i == 0:
                print(f"{n} is NOT prime.")
                return False
        print(f"{n} is PRIME.")
        return True

    def testPrims(self, a, b):
        def gcd(x, y):
            while y != 0:
                x, y = y, x % y
            return x

        if gcd(a, b) == 1:
            print(f"{a} and {b} are prime to each other.")
            return True
        else:
            print(f"{a} and {b} are NOT prime to each other.")
            return False

    def tableMult(self, n):
        print(f"Multiplication table of {n}:")
        for i in range(1, 11):
            print(f"{n} x {i} = {n * i}")

    def allTablesMult(self):
        for num in range(1, 10):
            self.tableMult(num)
            print()  # Line break

    @staticmethod
    def listDiv(n):
        Ldiv = []
        for i in range(1, n + 1):
            if n % i == 0:
                Ldiv.append(i)
        print(f"Divisors of {n} are: {Ldiv}")
        return Ldiv

    @staticmethod
    def listDivPrim(n):
        Ldiv = Computation.listDiv(n)
        LdivPrim = []
        for div in Ldiv:
            if Computation.isPrime(div):
                LdivPrim.append(div)
        print(f"Prime divisors of {n} are: {LdivPrim}")
        return LdivPrim

    @staticmethod
    def isPrime(n):
        if n <= 1:
            return False
        for i in range(2, int(n**0.5) + 1):
            if n % i == 0:
                return False
        return True
        # Class ka object banao
calc = Computation()

# Factorial
calc.Factorial(5)

# Natural Sum
calc.naturalSum(10)

# Test Prime
calc.testPrime(7)

# Test if 4 and 9 are prime to each other
calc.testPrims(4, 9)

# Multiplication Table of 5
calc.tableMult(5)

# All tables from 1 to 9
calc.allTablesMult()

# Divisors of 28
Computation.listDiv(28)

# Prime Divisors of 28
Computation.listDivPrim(28)

```

```python
class Instructor:
    def __init__(self):
        self.__name = None
        self.__technology_skill = []
        self.__experience = 0
        self.__average_feedback = 0.0

    # Setter methods
    def set_name(self, name):
        self.__name = name

    def set_technology_skill(self, skills):
        self.__technology_skill = skills  # skills should be a list

    def set_experience(self, experience):
        self.__experience = experience

    def set_average_feedback(self, avg_feedback):
        self.__average_feedback = avg_feedback

    # Eligibility check method
    def check_eligibility(self):
        if self.__experience > 3:
            return self.__average_feedback >= 4.5
        else:
            return self.__average_feedback >= 4.0

    # Allocate course method
    def allocate_course(self, technology):
        if self.check_eligibility() and technology in self.__technology_skill:
            return True
        else:
            return False

    # Display method for demo
    def display(self):
        print(f"Instructor Name: {self.__name}")
        print(f"Skills: {self.__technology_skill}")
        print(f"Experience: {self.__experience} years")
        print(f"Average Feedback: {self.__average_feedback}")

# Instructor 1
inst1 = Instructor()
inst1.set_name("Sudhanshu Singh")
inst1.set_technology_skill(["Python", "Java", "ML"])
inst1.set_experience(4)
inst1.set_average_feedback(4.6)

inst1.display()
print("Eligibility:", inst1.check_eligibility())
print("Course allocation (Python):", inst1.allocate_course("Python"))
print("Course allocation (C++):", inst1.allocate_course("C++"))

print("\n--------------------------------\n")

# Instructor 2
inst2 = Instructor()
inst2.set_name("Ravi Kumar")
inst2.set_technology_skill(["C++", "Data Structures"])
inst2.set_experience(2)
inst2.set_average_feedback(3.9)

inst2.display()
print("Eligibility:", inst2.check_eligibility())
print("Course allocation (C++):", inst2.allocate_course("C++"))
print("Course allocation (Python):", inst2.allocate_course("Python"))
```

```python
import random

class FlashCard:
    def __init__(self):
        # Fruits dictionary: fruit name = key, color = value
        self.fruits = {
            "Banana": "yellow",
            "Strawberries": "pink",
            "Apple": "red",
            "Grapes": "green",
            "Orange": "orange",
            "Blueberry": "blue",
            "Watermelon": "green"
        }

    def quiz(self):
        # Randomly select a fruit
        fruit = random.choice(list(self.fruits.keys()))
        color = self.fruits[fruit]

        print(f"What is the color of {fruit}?")
        user_answer = input("Your answer: ").strip().lower()

        if user_answer == color.lower():
            print("Correct! ✅")
        else:
            print(f"Wrong! ❌ The correct color is {color}.")
# FlashCard object banao
card = FlashCard()

# Quiz run karo
card.quiz()

```

```python
#Question:
#Ek Employee class_banao:

#Attributes: name, emp_id, basic_salary.

#Constructor: sab values lo.

#Method: calc_salary() — DA = 50% of basic, HRA = 20% of basic. Total salary nikalo aur return karo.

#Method: display() — details aur total salary print karo.

#int: Ye tumhe constructor + calculation + multiple methods ka proper feel dega
class Employee:
    def __init__(self, name, emp_id, basic_salary):
        self.name = name
        self.emp_id = emp_id
        self.basic_salary = basic_salary

    def calc_salary(self):
        da = 0.50 * self.basic_salary   # DA = 50% of basic
        hra = 0.20 * self.basic_salary  # HRA = 20% of basic
        total_salary = self.basic_salary + da + hra
        return total_salary

    def display(self):
        total_salary = self.calc_salary()
        print(f"Employee Name: {self.name}")
        print(f"Employee ID: {self.emp_id}")
        print(f"Basic Salary: ₹{self.basic_salary}")
        print(f"Total Salary (with DA & HRA): ₹{total_salary}")

# Employee object banao
e1 = Employee("Sudhanshu", 101, 50000)

# Display details aur total salary
e1.display()

print("\n---\n")

e2 = Employee("Amit", 102, 30000)
e2.display()
```

```python
#Q5. Circle Class with Operator Overloading

#Circle class banao jisme radius ho.

#Constructor se radius lo.

#Method: area(), circumference().

#__add__ overload karo — agar do Circle objects add karo toh naya Circle banao jiska radius = dono ka sum.
class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius

    def circumference(self):
        return 2 * 3.14 * self.radius

    def __add__(self, other):
        new_radius = self.radius + other.radius
        return Circle(new_radius)
c1 = Circle(5)
c2 = Circle(3)

print(f"Circle 1 - Radius: {c1.radius}, Area: {c1.area()}, Circumference: {c1.circumference()}")
print(f"Circle 2 - Radius: {c2.radius}, Area: {c2.area()}, Circumference: {c2.circumference()}")

print("\n--- Adding circles ---\n")

c3 = c1 + c2   # __add__ call hoga

print(f"New Circle - Radius: {c3.radius}, Area: {c3.area()}, Circumference: {c3.circumference()}")
```
